# Final Load and Preparation - House Price Prediction

This notebook performs the final data preparation and loading steps to ensure the dataset is ready for visualization in Tableau and any final reporting.

```python
import pandas as pd
import numpy as np
import os

print("Libraries imported.")
```

## 1. Load Cleaned Data

```python
data_path = '../data/processed/cleaned_data.csv'
df = pd.read_csv(data_path)
print(f"Loaded dataset with {df.shape[0]} rows and {df.shape[1]} columns.")
```

## 2. Final Data Type Consolidation
Ensure that columns like 'YearBuilt' and 'MSSubClass' are treated correctly for analysis.

```python
# Convert years to integers if they aren't already
year_cols = [col for col in df.columns if 'Year' in col]
for col in year_cols:
    df[col] = df[col].astype(int)

print("Data types consolidated.")
```

## 3. Preparation for Tableau
Adding descriptive columns or labels if needed for better visualization.

```python
# Example: Create a 'Price Category' column using the correct column name 'Saleprice'
df['PriceCategory'] = pd.qcut(df['Saleprice'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

print("Final preparation complete.")
```

## 4. Export Final Tableau Dataset

```python
output_path = '../data/processed/tableau_ready.csv'
df.to_csv(output_path, index=False)
print(f"Final dataset exported to: {output_path}")
```
